# Overfitting, Underfitting & Good Fit

```python
Real Life Analogy First 📖
Imagine a student preparing for exams:

UNDERFITTING student:
→ studied too little
→ doesn't know anything
→ fails in practice tests
→ fails in real exam too
→ "too simple, learned nothing!"

OVERFITTING student:
→ memorized every practice question
→ scores 100% in practice tests
→ real exam has new questions
→ fails badly in real exam!
→ "memorized, didn't understand!"

GOOD FIT student:
→ understood the concepts
→ scores well in practice tests
→ scores well in real exam too
→ "learned the pattern!" ✅
```

```python
In ML Terms:

UNDERFITTING:
→ model is too simple
→ bad on train data
→ bad on test data
→ "model didn't learn enough!"

train accuracy = 55%
test accuracy  = 53%
both are bad! ❌

OVERFITTING:
→ model is too complex
→ great on train data
→ bad on test data
→ "model memorized train data!"

train accuracy = 99%
test accuracy  = 60%
big gap! ❌

GOOD FIT:
→ model learned patterns
→ good on train data
→ good on test data

train accuracy = 90%
test accuracy  = 87%
small gap! ✅
```

```python
Visual Explanation:

UNDERFITTING:
Data points:  * * * * * *
Model line:   ___________  (flat, too simple!)
Model missed the pattern completely!

GOOD FIT:
Data points:  * * * * * *
Model line:   ___/‾‾‾‾‾  (captures trend!)
Model learned the general pattern! ✅

OVERFITTING:
Data points:  * * * * * *
Model line:   _/\_/‾\_/‾  (wiggly, too complex!)
Model memorized every point!
Fails on new data!
```

```python
In Decision Tree specifically:


UNDERFITTING:
max_depth = 1
→ only 1 question
→ too simple
→ misses many patterns

         [age > 40?]
         /         \
   Not Bought     Bought
   
   Too simple! Many wrong predictions!

OVERFITTING:
max_depth = None
→ tree keeps splitting
→ memorizes every sample

         [age > 40?]
         /         \
   [salary>20?]  [salary>80?]
    /      \       /       \
[age>22?] ...  [age>55?]  ...
  /    \          /    \
 ...   ...      ...    ...
 
 Too complex! Fails on new data!

GOOD FIT:
max_depth = 3
→ learns main patterns
→ generalizes well ✅
```

# How to Fix Overfitting:

```python
Decision Tree:
→ reduce max_depth ✅
→ increase min_samples_split ✅
→ increase min_samples_leaf ✅
→ use Random Forest (next algo!) ✅

General ML:
→ get more training data ✅
→ remove useless features ✅
→ use simpler model ✅
→ use regularization ✅
```

# How to Fix Underfitting:

```python
→ increase max_depth ✅
→ use more features ✅
→ use complex model ✅
→ train for longer ✅
→ reduce regularization ✅
```

```python
The Goldilocks Zone 🎯


          Underfitting    Good Fit    Overfitting
               ❌            ✅            ❌
               
Complexity: too simple → just right → too complex
max_depth:      1      →     3      →    None
Train acc:     low     →    high    →   very high
Test acc:      low     →    high    →     low

Goal: find the JUST RIGHT complexity!
Like Goldilocks — not too hot, not too cold! 😄
```

```python
Notes Summary 📝

Underfitting:
→ model too simple
→ bad train AND test accuracy
→ fix: increase complexity

Overfitting:
→ model too complex
→ great train, bad test accuracy
→ fix: reduce complexity

Good Fit:
→ just right!
→ good train AND test accuracy
→ small gap between them

How to detect:
→ always check BOTH train and test accuracy!
→ big gap = overfitting
→ both low = underfitting

Golden Rule:
train accuracy - test accuracy < 5% = good! ✅
```
