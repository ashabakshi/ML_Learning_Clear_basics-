# Evaluation Metrics for Classification

Accuracy 

```python
most basic metric

accuracy = correct predictions
           ──────────────────
            total predictions

if 90 out of 100 correct → accuracy = 90%
```

Confusion Matrix:

```python
Predicted
              Fail(0)  Pass(1)
Actual Fail(0) [ TN      FP  ]
Actual Pass(1) [ FN      TP  ]

TN = True Negative  → predicted Fail,  actual Fail  ✅
TP = True Positive  → predicted Pass,  actual Pass  ✅
FP = False Positive → predicted Pass,  actual Fail  ❌
FN = False Negative → predicted Fail,  actual Pass  ❌
```

Classification Report:

```python
Precision = out of all predicted Pass
            how many actually passed?

Recall    = out of all actual Pass
            how many did model catch?

F1 Score  = balance between precision and recall
            best metric when data is imbalanced!
```
