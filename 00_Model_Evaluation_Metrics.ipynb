# Model Evaluation Metrics — Complete Guide

1. MSE — Mean Squared Error

What it does:

```python

Measures average squared difference
between actual and predicted values

MSE = Σ(actual - predicted)²
      ──────────────────────
             n

Your output:
actual    = [90, 40]   (2 test samples)
predicted = [92, 41]
errors    = [-2, -1]
squared   = [4,   1]
MSE       = (4+1)/2 = 2.5
When to use:
✅ when you want to penalize large errors heavily
✅ used internally by model during training
✅ comparing two models on same dataset

When NOT to use:

❌ when explaining to non-technical people
   "MSE is 45,000,000" → nobody understands this!
❌ when you want error in original units


Problem with MSE:

salary data → error = 3000
MSE = 3000² = 9,000,000

Very large number!
Hard to understand!
Units are squared (₹²) → makes no sense!
```

2. RMSE — Root Mean Squared Error

What it does:

```python
Simply √MSE
Brings error back to original units!

RMSE = √MSE

If MSE = 9,000,000
RMSE = √9,000,000 = 3000

Now you can say:

"model is off by ₹3000 on average" ✅
Easy to understand!
When to use:
✅ when you want error in original units
✅ when explaining to business/non-technical people
✅ when large errors are very bad
   (RMSE punishes large errors more than small ones)

When NOT to use:

❌ when your data has many outliers
   (outliers make RMSE very large)
   use MAE instead!

Punishes large errors:

Error of 10 → 10² = 100
Error of 20 → 20² = 400  (4x penalty for 2x error!)

So RMSE is strict about big mistakes!
```

3. MAE — Mean Absolute Error

What it does:

```python
Average of absolute errors
(no squaring!)

MAE = Σ|actual - predicted|
      ──────────────────────
              n

errors    = [-2, 5, -3, 4]
absolute  = [2,  5,  3, 4]
MAE       = (2+5+3+4)/4 = 3.5

When to use:
✅ when data has outliers
✅ when all errors are equally important
✅ easier to explain than RMSE

When NOT to use:

❌ when large errors are very costly
   (MAE treats all errors equally)
   use RMSE instead!
MAE vs RMSE:
Data: actual=[100], predicted=[200]
error = 100

MAE  = 100        (treats it normally)
RMSE = 100        (same for single point)

Data: actual=[100,100], predicted=[200, 100]
errors = [100, 0]

MAE  = (100+0)/2  = 50   (fair average)
RMSE = √(10000/2) = 70.7 (punishes the big error more!)

Use MAE  → when outliers exist, treat fairly
Use RMSE → when big errors are unacceptable!
```

4. R2 Score — Coefficient of Determination

What it does:

```python
Tells how much of the pattern
model captured from data

R2 = 1 - (MSE of model)
         ───────────────
         (MSE of baseline)

baseline = just predicting mean every time!

R2 = 1   → perfect model 🎯
R2 = 0   → model = just predicting mean (useless!)
R2 < 0   → model worse than mean! something wrong!
Simple explanation:
Imagine class average marks = 70

Baseline model → always predicts 70 for everyone
Your model     → predicts based on study hours

R2 = 0.99 means:

"your model is 99% better than just predicting mean!"

When to use:

✅ most common metric for regression
✅ easy to understand (percentage scale)
✅ comparing different models
✅ always use this first!

When NOT to use:

❌ when comparing models with different features
   use Adjusted R2 instead!
❌ R2 always increases when you add more features
   even useless features!
   this is misleading!

   
Scale to remember:
R2 >= 0.90 → Excellent 🎯
R2 >= 0.80 → Good ✅
R2 >= 0.60 → Average 😐
R2 >= 0.40 → Poor ❌
R2 <  0.40 → Very Poor 💀
```

5. Adjusted R2 — The Smarter R2!

The problem with R2:

```python
Every time you add a new feature
R2 ALWAYS increases or stays same
even if feature is useless!

Example:
Features: study_hours → R2 = 0.85
Add sleep_hours       → R2 = 0.87  (makes sense!)
Add student's name    → R2 = 0.86  (name is useless!)

R2 never punishes useless features!
Misleading! ❌


Adjusted R2 fixes this:
Adjusted R2 penalizes useless features!

Add useful feature   → Adjusted R2 increases ✅
Add useless feature  → Adjusted R2 decreases ❌

Formula:
Adj R2 = 1 - (1-R2)(n-1)
              ──────────
               (n-p-1)

n = number of samples
p = number of features
Code:
python# sklearn doesn't have adjusted r2
# calculate manually!

r2 = r2_score(y_test, y_pred)
n  = len(y_test)      # number of samples
p  = X_test.shape[1]  # number of features

adj_r2 = 1 - (1-r2) * (n-1) / (n-p-1)
print("Adjusted R2:", round(adj_r2, 4))

When to use:

✅ when you have multiple features
✅ when comparing models with different
   number of features
✅ always prefer over R2 in real projects!

When NOT to use:

❌ when you have only 1 feature
   (R2 and Adjusted R2 will be same!)
```

```python
Complete Comparison Table 📊

Metric      Unit          Best Value   Penalizes    Use When
────────────────────────────────────────────────────────────
MSE         squared       0            big errors   internal
RMSE        original      0            big errors   reporting
MAE         original      0            equally      outliers
R2          none (0-1)    1            nothing      always
Adj R2      none (0-1)    1            extra feat   multiple X
```

```python
Which metric to use when? 🎯

Single feature model?
→ R2 + RMSE  ✅

Multiple feature model?
→ Adjusted R2 + RMSE  ✅

Data has outliers?
→ MAE instead of RMSE  ✅

Explaining to business?
→ RMSE or MAE  ✅
  ("model is off by ₹3000 on average")

Comparing two models?
→ R2 + RMSE together  ✅
```

```python
from sklearn.metrics import mean_absolute_error

mse    = mean_squared_error(y_test, y_pred)
rmse   = np.sqrt(mse)
mae    = mean_absolute_error(y_test, y_pred)
r2     = r2_score(y_test, y_pred)

n      = len(y_test)
p      = X_test.shape[1]
adj_r2 = 1 - (1-r2) * (n-1) / (n-p-1)

print("MSE        :", round(mse, 2))
print("RMSE       :", round(rmse, 2))
print("MAE        :", round(mae, 2))
print("R2         :", round(r2, 4))
print("Adjusted R2:", round(adj_r2, 4))
```

```python
Notes Summary 📝

MSE   → squared error, used internally, hard to read
RMSE  → √MSE, same unit as target, punishes big errors
MAE   → absolute error, fair to outliers
R2    → % of pattern captured, always use this
Adj R2→ smarter R2, penalizes useless features

Golden Rule:
Always report R2 + RMSE together!
If multiple features → use Adjusted R2!
If outliers → use MAE instead of RMSE.
```
